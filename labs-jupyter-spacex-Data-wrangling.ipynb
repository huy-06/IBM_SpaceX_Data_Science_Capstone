```python
import pandas as pd
import numpy as np

# 1. Đọc dữ liệu từ file dataset_part_1.csv đã tạo ở Bước 1
df = pd.read_csv("dataset_part_1.csv")

# Xem qua 5 dòng đầu tiên
print("--- 5 dòng dữ liệu đầu tiên ---")
print(df.head(5))

# 2. Xác định phần trăm dữ liệu bị khuyết (NaN) ở các cột
print("\n--- Phần trăm dữ liệu bị thiếu ở mỗi cột ---")
print(df.isnull().sum() / len(df) * 100)

# Xem các loại dữ liệu của từng cột
print("\n--- Kiểu dữ liệu các cột ---")
print(df.dtypes)

# 3. Phân tích số lần phóng ở các bệ phóng (Launch Site)
print("\n--- Số lần phóng tại mỗi địa điểm phóng (Launch Site) ---")
launch_outcomes = df['LaunchSite'].value_counts()
print(launch_outcomes)

# 4. Phân tích các loại Quỹ đạo (Orbit)
print("\n--- Số lần phóng của từng loại Quỹ đạo (Orbit) ---")
orbit_counts = df['Orbit'].value_counts()
print(orbit_counts)

# 5. Phân tích kết quả hạ cánh tên lửa (Outcome)
print("\n--- Các loại kết quả hạ cánh (Outcome) và số lượng ---")
landing_outcomes = df['Outcome'].value_counts()
print(landing_outcomes)

# Trích xuất danh sách các kết quả thất bại (hoặc không hạ cánh)
# landing_outcomes.keys() cho ta các giá trị độc nhất trong cột Outcome
for i, outcome in enumerate(landing_outcomes.keys()):
    print(f"{i}: {outcome}")

# 6. Tạo danh sách các kết quả hạ cánh THÀNH CÔNG
# Các trạng thái bắt đầu bằng "True" là thành công
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]]) # Các kết quả thất bại/không hạ cánh
print("\nCác kết quả hạ cánh thất bại (bad outcomes):", bad_outcomes)

# 7. Tạo cột 'Class' (Nhãn mục tiêu): 1 = Thành công, 0 = Thất bại
# Sử dụng list comprehension để gán nhãn
landing_class = []
for outcome in df['Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

df['Class'] = landing_class

# Kiểm tra tỷ lệ thành công trung bình của các lần phóng
success_rate = df['Class'].mean()
print(f"\nTỷ lệ phóng và hạ cánh thành công trung bình: {success_rate * 100:.2f}%")

# 8. Xem lại DataFrame sau khi đã có thêm cột Class
print("\n--- 5 dòng đầu tiên kèm nhãn Class mới ---")
print(df[['FlightNumber', 'Date', 'Outcome', 'Class']].head(5))

# 9. Lưu dữ liệu sạch thành file dataset_part_2.csv
df.to_csv("dataset_part_2.csv", index=False)
print("\nXử lý dữ liệu hoàn tất! File 'dataset_part_2.csv' đã được lưu thành công.")
```
